<h1 style="text-align: center; font-weight: bold;">Visualizing K means algorithm</h1>

K-means algorithm is a widely used clustering algorithm in machine learning. It is an unsupervised learning algorithm that groups data points into k clusters based on their similarity. K-means algorithm is an iterative algorithm that optimizes the cluster assignments by minimizing the sum of squared distances between the data points and their assigned cluster centroids. In this project, we will explore how to visualize the K-means algorithm using Python's data visualization libraries.

I implemented the visualizer through modifying `KMeans.cluster()` to include boolean parameter `animate`. When animate is set to True, the algorithm will perform kmeans clustering while also displaying each step on the plot. We can then observe the change in centroids and cluster labels of data

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import kmeans
```

we first plot the five_blobs data

```python
five_blobs = pd.read_csv('./data/five_blobs.csv').to_numpy()

plt.scatter(five_blobs[:, 0], five_blobs[:, 1])
plt.title('five_blobs.csv')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
np.random.seed(0)
ani_cluster = kmeans.KMeans(five_blobs)
```

```python
%matplotlib notebook
from IPython.display import HTML

inertia, step, ani = ani_cluster.cluster(5, animate=True)

plt.show()
```

As we can see, the program run and let us observe each step it took to fully classify the clusters in the dataset.
